```
#subset_df = df[0:5]
#subset_prev = df_prev[0:5]
#subset_bureau = df_bureau[0:5]

#es = ft.EntitySet(id = 'clients')

#es = es.entity_from_dataframe(entity_id = 'app', dataframe = subset_df, index = 'SK_ID_CURR')
#es = es.entity_from_dataframe(entity_id = 'previous', dataframe = subset_prev, index = 'SK_ID_PREV')
#es = es.entity_from_dataframe(entity_id = 'bureau', dataframe = subset_bureau, index = 'SK_ID_BUREAU')

es = ft.EntitySet(id = 'clients')

es = es.entity_from_dataframe(entity_id = 'app', dataframe = df, index = 'SK_ID_CURR')
es = es.entity_from_dataframe(entity_id = 'previous', dataframe = df_prev, index = 'SK_ID_PREV')
es = es.entity_from_dataframe(entity_id = 'bureau', dataframe = df_bureau, index = 'SK_ID_BUREAU')


r_app_previous = ft.Relationship(es['app']['SK_ID_CURR'], es['previous']['SK_ID_CURR'])
r_app_bureau = ft.Relationship(es['app']['SK_ID_CURR'], es['bureau']['SK_ID_CURR'])


es = es.add_relationships([ r_app_previous,r_app_bureau])

print(es)                  
```

```
# Default primitives from featuretools
default_agg_primitives =  ["sum", "std", "max", "skew", "min", "mean", "count", "mode","n_most_common","median"]
#default_trans_primitives =  ["add_numeric", "multiply_numeric"]

default_trans_primitives = ["percentile"]
#default_trans_primitives = ["greater_than","less_than"]


# DFS with specified primitives
feature_matrix, feature_names = ft.dfs(entityset = es, target_entity = 'app',
#                       trans_primitives = default_trans_primitives,
                       agg_primitives=default_agg_primitives,
#                      primitive_options={
#                                       ('greater_than', 'less_than'): {
#                                        'include_variables': {'app': numerical_principal_reduced}},
              #                           ("max", "min", "mean"): {
               #                         'include_variables': {'app': numerical_principal_reduced}}                                   
                                        
#                                       }, 
                       max_depth = 2, features_only=False, verbose = True)
                       
# Reset the index to make SK_ID_CURR a column again                                      
feature_matrix = feature_matrix.reset_index()
```
