### Construção da matriz de features
- Primitives :Funções que serão aplicadas na criação da matrix de features 
- Com a combinação de no máximo 2 features (max_depth =2)

```
# Default primitives from featuretools
default_agg_primitives =  ["sum", "std", "max", "skew", "min", "mean", "count", "num_unique", "mode"]
default_trans_primitives =  ['add_numeric', 'multiply_numeric']

feature_matrix, feature_names = ft.dfs(entityset = es, target_entity = 'app',
                                      trans_primitives = default_trans_primitives,
                                       #agg_primitives=default_agg_primitives, 
                                        max_depth = 2, 
                                       primitive_options={
                                        ('add_numeric', 'multiply_numeric'): {
                                        'include_variables': {'app': principal}},                                      
                                        
                                        },
                                       
                                        verbose = True)

feature_matrix = feature_matrix.reset_index()

pd.options.display.max_columns = 1700
feature_matrix.head(10)
```
